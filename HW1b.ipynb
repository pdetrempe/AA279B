```julia
using LinearAlgebra
using Plots
using DifferentialEquations
```

```julia
μₑ = 398600 # [km^3/s^2]
μₘ = 4903
R = 384400 # Earth-Moon distance [km]

R_Earth = 6378 # Earth radius [km]
R_Moon = 1738  # Moon radius  [km]
h = 200

μ₁ = μₑ
μ₂ = μₘ

# Synodic reference frame parameters
R₁ = μ₂/(μ₂+μ₁)*R   # Earth-CM distance
R₂ = μ₁/(μ₂+μ₁)*R        # Moon-CM distance

# Initial conditions
v_inj = 10.92367104     # [km/s]
ϕ_inj = deg2rad(47.70061087)

# Equations of motion
r1 = [-R₁; 0; 0]
r2 = [ R₂; 0; 0]
ωₛ = [0;0;sqrt((μ₁ + μ₂)/(R^3))]

# Initial conditions (synodic)
r0 = [-cos(ϕ_inj)*(h+R_Earth)-R₁; -sin(ϕ_inj)*(h+R_Earth); 0] # No z-position/velocity
v0 = [v_inj*sin(ϕ_inj); -v_inj*cos(ϕ_inj); 0]
x0 = [r0; v0]
```

```julia
function CR3BP_EOM(x, p, t)
    μ₁,μ₂,r1,r2,ωₛ = p
    r3 = x[1:3]
    velocity = x[4:6]

    r13 = r3-r1
    r23 = r3-r2

    acceleration = -μ₁*r13/norm(r13)^3 -μ₂*r23/norm(r23)^3 -    # gravity
                    cross(ωₛ, cross(ωₛ, r3)) -                  # centripetal
                    2*cross(ωₛ, velocity)                       # coriolis
                   # ^gravity, centripetal, coriolis

    x_dot = [velocity; acceleration]
end


```

```julia
tspan = (0.,8*24*3600)

# Integrate trajectory
params = (μ₁,μ₂,r1,r2,ωₛ)
free_return = ODEProblem(CR3BP_EOM,x0,tspan,params)
sol = solve(free_return,alg_hints=[:stiff],reltol=1e-8, VCABM())
```

```julia
# Plot trajectory in synodic frame
gr()
plot(sol,vars=(1,2),denseplot=false)
```

```julia
# 1d) Rotate to inertial coordinates
θ = ωₛ[3].*sol.t
function state_synodicToInertial(state_syn,θ)
    pos_syn = state_syn[1:3]
    vel_syn = state_syn[4:6]
    
    # Rotate position
    T = [cos(θ) -sin(θ) 0; sin(θ) cos(θ) 0; 0 0 1]
    pos_inert = T*pos_syn
    
    # Transform velocity (taking rotating frame into account)
    vel_inert = T*(vel_syn + cross(ωₛ, pos_syn))
    
    return [pos_inert; vel_inert]
end

state_inert = state_synodicToInertial.(sol.u, θ)
```

```julia
# Manipulate data and plot in inertial frame
state_Arr = zeros(6, length(θ))
for (idx, state) in enumerate(state_inert)
    state_Arr[:, idx] = state
end
plot(state_Arr[1,:], state_Arr[2,:])
```

```julia
# 1e) Plot specific orbital energy/Jacobi constant vs. time
function E_effective(state)
    r3 = state[1:3]
    v = state[4:6]
    r13 = r3-r1
    r23 = r3-r2

    ε_eff = -μ₁/norm(r13) -μ₂/norm(r23) -
             1/2*ωₛ[3]^2*(r3[1]^2+r3[2]^2) +       # Rotational kinetic energy 
             1/2*norm(v)^2                         # Translational kinetic energy
    C = -2*ε_eff # Jacobi constant
    
    return ε_eff, C
end
ε_eff = E_effective.(sol.u)
```

```julia
# 1f) Plot specific orbital energy (need inertial velocity)
function CR3BP_OrbitalEnergy(state_synodic, state_inertial) # Use both states because r1 and r2 defined in synodic frame
    r3 = state_synodic[1:3]
    v = state_inertial[4:6]
    r13 = r3-r1
    r23 = r3-r2
    
    ε = -μ₁/norm(r13) -μ₂/norm(r23) + 1/2*norm(v)^2
    return ε
end
ε = CR3BP_OrbitalEnergy.(sol.u, state_inert)
plot(ε)
```

```julia
# 1g) Integrate trajectory in inertial coordinates (assuming Moon moves on circle)
rotateVector(θ, vec) = [cos(θ) -sin(θ) 0; sin(θ) cos(θ) 0; 0 0 1]*vec


function CR3BP_EOM_inertial(state_inertial, params, t)
    μ₁,μ₂,r1,r2,ωₛ = params

    r3 = state_inertial[1:3]
    v  = state_inertial[4:6]
    r_Moon = rotateVector(ωₛ[3]*t, r2)
    r_Earth = rotateVector(ωₛ[3]*t, r1)
    r13 = r3-r_Earth
    r23 = r3-r_Moon
    
    acceleration = -μ₁*r13/norm(r13)^3 -μ₂*r23/norm(r23)^3
    
    xdot = [v; acceleration]
end

```

```julia
tspan = (0.,6*24*3600)

# Integrate trajectory
params = (μ₁,μ₂,r1,r2,ωₛ)
x0_inert = state_synodicToInertial([r0;v0],0)
free_return_inertial = ODEProblem(CR3BP_EOM_inertial, x0_inert, tspan, params)
sol2 = solve(free_return_inertial,alg_hints=[:stiff],reltol=1e-8, VCABM())

# Plot inertially integrated trajectory
gr()
plot(sol2,vars=(1,2),denseplot=false)
```
